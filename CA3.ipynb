```python
import pandas as pd
sep = 72*'-'

# the class to process changes
class ChangeProcessor(object):

    def __init__(self, file_name):
        self.read_file(file_name)

    # open the file - and read all of the lines.
    # use strip to strip out spaces and trim the line.
    def read_file(self, file_name):
        self.data = [line.strip() for line in open(file_name, 'r')]
        self.process_commits()
        return len(self.data)

    def process_commits(self):
        self.commits = []
        current_commit = None
        index = 0

        while True:
            try:
                # parse each of the commits and put them into a list of commits
                details = self.data[index + 1].split('|')
                # the author with spaces at end removed.

                # add details to the list of commits.
                self.commits.append(details)
                index = self.data.index(sep, index + 1)

            except IndexError:
                break
            #except ValueError:
            #    break
        return self.commits

    def get_author(self, index):
		return self.commits[index][1].strip()

    def get_date(self, index):
		return self.commits[index][2].strip()

    def get_number_of_comment_lines(self, index):
		return self.commits[index][3].strip()

    def get_revision(self, index):
		return self.commits[index][0].strip()

    def get_number_of_commits(self):
        return len(self.commits)

change = ChangeProcessor('changes_python.txt')
limit = change.get_number_of_commits()
authors = [change.get_author(i) for i in range(limit)]
dates = [change.get_date(i) for i in range(limit)]
size = [change.get_number_of_comment_lines(i) for i in range(limit)]
reviz = [change.get_revision(i) for i in range(limit)]

data = pd.DataFrame({'authors':authors,'dates':dates,'length':size,'revision type':reviz})
print data

```
